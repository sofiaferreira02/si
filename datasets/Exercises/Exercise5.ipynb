##### 5.2) Test the PCA class in a jupyter notebook using the iris.csv dataset (classificação).

```python
from si.io.csv_file import read_csv
from si.decomposition.pca import PCA as PCA
```

```python
df_iris = read_csv(r'C:/Users/35192/si/datasets/iris/iris.csv', sep=",", label=True)

print("Dataset has label?:",df_iris.has_label())
print("Label:", df_iris.label)
print("Label classes:",df_iris.get_classes())
print("Features:", df_iris.features)
print("Shape of dataset:", df_iris.shape())
```

```python
# PCA with 3 components
pca = PCA(n_components=3)
pca.fit(df_iris)
```

```python
# Transform the dataset
iris = pca.transform(df_iris)

print(f"Principal Components (Eigenvectors): \n{pca.components}")
print(f"\nExplained Variance (Proportion): \n{pca.explained_variance}")
print(f"\nReduced Dataset (first 5 samples): \n{iris.X[:5]}")
print(f"\nFeature Names of the Reduced Dataset: \n{iris.features}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

label_map = {label: idx for idx, label in enumerate(iris.get_classes())}
num_labels = np.array([label_map[label] for label in iris.y])


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(iris.X[:, 0], iris.X[:, 1], iris.X[:, 2], c=num_labels, cmap='coolwarm', edgecolors='blue')
ax.set_xlabel('PCA 1')
ax.set_ylabel('PCA 2')
ax.set_zlabel('PCA 3')
ax.set_title('PCA 3D - Iris ')
colorbar = fig.colorbar(scatter, ax=ax)
ax.grid(alpha=0.25)
plt.show()
```
